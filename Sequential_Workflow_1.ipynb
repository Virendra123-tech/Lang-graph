```python
from datetime import datetime
```

```python
def add_time(state:Greeting):
    date_time=datetime.now().strftime("%H:%M:%S")
    state["final_message"]=f" {state['gretting']} Current time is {date_time}"
    return state
```

## Data Cleaning Pipeline

```python
from typing import List,Dict
```

```python
class Data_cleaning(TypedDict):
    data:list[Dict]
    cleaned_data:list[Dict]
```

```python
def Remove_nullvalues(state:Data_cleaning):
    cleaned_data=[]
    for row in state['data']:
        new_row={k:v for k,v in row.items() if v not in [None,""]}
        cleaned_data.append(new_row)
        state['cleaned_data']=cleaned_data
    return state
        
        
```

```python
def normalized_date(state:Data_cleaning):
    normalized=[]
    for row in  state['cleaned_data']:
        new_row={}
        for k,v in row.items():
            if isinstance(v,str):
                new_row[k]=v.lower().strip()
            else:
                new_row[k]=v
            normalized.append(new_row)
        state['cleaned_data']=normalized
        return state
```

```python
graph=StateGraph(Data_cleaning)

graph.add_node('Remove_nullvalues',Remove_nullvalues)
graph.add_node('normalized_date',normalized_date)

# Edge 
graph.add_edge(START,'Remove_nullvalues')
graph.add_edge('Remove_nullvalues','normalized_date')
graph.add_edge('normalized_date',END)
workflow=graph.compile()
```
